```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
```

```python
# Set up Chrome options to force download PDF files
chrome_options = Options()
chrome_options.add_experimental_option('prefs', {
    "download.prompt_for_download": False,
    "download.default_directory": "/Users/seyfal/Desktop/download_papers",  # Change to your download directory
    "plugins.always_open_pdf_externally": True,  # This will force download PDFs
    "safebrowsing.enabled": True
})

# Path to the ChromeDriver
chrome_driver_path = 'your/path'  # Change to the path where you have chromedriver

# Set up WebDriver
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Credentials
email = 'your@email.com'  # Replace with your email
password = 'your_password'  # Replace with your password

# URL of the webpage
base_url = 'your_link'  # Replace with the URL of the webpage
```

```python
try:
    # Open the login page
    driver.get(base_url.format(0))  # Start from page 0
    
    time.sleep(5)  # Adjust sleep time as needed
    
    # Locate and fill in the email field
    email_field = driver.find_element(By.ID, 'i0116')
    email_field.send_keys(email)
    
    # Submit the email and wait for the password field to load
    email_field.send_keys(Keys.RETURN)
    time.sleep(3)  # Adjust sleep time as needed
    
    # Locate and fill in the password field
    password_field = driver.find_element(By.ID, 'passwordInput')
    password_field.send_keys(password)
    
    # Submit the password and wait for login to complete
    password_field.send_keys(Keys.RETURN)
    time.sleep(5)  # Adjust sleep time as needed
    
    # Loop through pages
    while True:
        # Locate PDF download buttons
        pdf_buttons = driver.find_elements(By.CLASS_NAME, 'icon-pdf-file')
        
        print(f"Found {len(pdf_buttons)} PDF buttons on the page.")
        
        for index, button in enumerate(pdf_buttons):
            try:
                # Find the parent link of the PDF icon
                parent = button.find_element(By.XPATH, '..')
                
                # Get the href attribute of the parent link
                link = parent.get_attribute('href')
                
                print(f"Processing link {index + 1}/{len(pdf_buttons)}: {link}")
                
                # Use JavaScript to open the link in a new tab
                driver.execute_script("window.open(arguments[0]);", link)
                
                # Switch to the new tab
                driver.switch_to.window(driver.window_handles[-1])
                
                # Wait for the PDF to load and start downloading
                time.sleep(1)  # Adjust sleep time as needed
                
                # Switch back to the original tab
                driver.switch_to.window(driver.window_handles[0])
                
            except Exception as e:
                print(f"Failed to download PDF {index + 1}/{len(pdf_buttons)}: {e}")
        
        # Find the "Next Page" button and click on it
        next_page_button = driver.find_element(By.CLASS_NAME, 'pagination__btn--next')
        if next_page_button.get_attribute('href'):
            next_page_button.click()
        else:
            print("No more pages to navigate to.")
            break
        
        time.sleep(3)  # Adjust sleep time as needed for the next page to load

finally:
    # Close the WebDriver
    driver.quit()
```
